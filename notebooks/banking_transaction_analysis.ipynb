# Banking Transaction Analysis
## Objective
The objective of this project is to analyze banking transaction data to identify revenue trends, customer spending behavior, demographic insights, and high-value transaction patterns..

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv(r"bank_transactions_clean.csv")

df.head()
```

## 1. Monthly Revenue Trend

```python
df["Month"] = df["TransactionDate"].dt.to_period("M")

monthly_revenue = df.groupby("Month")["TransactionAmount"].sum()

plt.figure()
monthly_revenue.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

## 2. Gender-wise Total Spending

```python
gender_revenue = df.groupby("CustGender")["TransactionAmount"].sum()

plt.figure()
gender_revenue.plot(kind='bar')
plt.title("Total Transaction Amount by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Revenue")
plt.show()
```

## 3. Top 10 Locations by Revenue

```python


location_revenue = df.groupby("CustLocation")["TransactionAmount"].sum()

plt.figure()
location_revenue.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Locations by Total Revenue")
plt.xlabel("Location")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

## 4. Transaction Amount Distribution

```python


plt.figure()
plt.boxplot(df["TransactionAmount"])
plt.title("Transaction Amount Distribution")
plt.ylabel("Transaction Amount")
plt.show()
```

## 5. Top 10 Customers by Spending

```python

top_customers = df.groupby("CustomerID")["TransactionAmount"].sum()

plt.figure()
top_customers.sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Customers by Total Spending")
plt.xlabel("Customer ID")
plt.ylabel("Total Spending")
plt.xticks(rotation=45)
plt.show()
```

## Key Insights

### - Revenue shows monthly variation indicating seasonal spending patterns.
### - Certain locations contribute significantly higher transaction volumes.
### - A small group of customers accounts for a large share of total revenue.
### - Age group 26-35 contributes the highest spending.
### - Presence of high-value transactions suggests potential premium customers or risk monitoring needs.
